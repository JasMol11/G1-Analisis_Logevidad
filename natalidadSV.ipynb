ORGANIZACIÓN MUNDIA DE LA SALUD: Life expectancy at birth
The average number of years that a newborn could expect to live. El Salvador, by sex, 2000 - 2021.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Importación del archivo CSV
archivo_csv = "Natalidad.csv"  # Nombre del archivo
data = pd.read_csv(archivo_csv)

# Configuración del gráfico
plt.figure(figsize=(12, 8))

# Graficar esperanza de vida para mujeres, hombres y general
plt.plot(data["Año"], data["Esperanza de vida-Mujeres"], label="Mujeres", marker='o')
plt.plot(data["Año"], data["Esperanza de vida-Hombres"], label="Hombres", marker='s')
plt.plot(data["Año"], data["Esperanza de vida"], label="General", marker='^')

# Personalización del gráfico
plt.title("Evolución de la Esperanza de Vida (1960-2022)", fontsize=16)
plt.xlabel("Año", fontsize=12)
plt.ylabel("Esperanza de vida (años)", fontsize=12)
plt.legend()
plt.grid(True)


# Guardar la gráfica como imagen
plt.savefig("esperanza_vida.png", format="png", dpi=300)
print("Gráfica guardada como esperanza_vida.png")

# Mostrar la gráfica
plt.show()

```
