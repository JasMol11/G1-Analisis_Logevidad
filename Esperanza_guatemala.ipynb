```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de esperanza de vida
datos = [
    45.808, 46.47, 46.995, 47.407, 47.712, 47.814, 47.764, 47.984, 48.298, 48.548,
    49.439, 50.484, 51.638, 53.003, 54.039, 54.824, 49.169, 55.816, 55.963, 56.185,
    56.024, 55.972, 50.134, 58.718, 59.558, 60.203, 60.772, 61.432, 62.029, 62.374,
    62.473, 62.725, 62.913, 63.349, 63.726, 64.195, 64.863, 65.536, 66.114, 66.869,
    67.446, 67.946, 68.386, 68.797, 69.165, 69.168, 69.812, 70.1, 70.356, 70.641,
    70.878, 71.177, 71.463, 71.728, 71.964, 72.103, 72.355, 72.551, 72.726, 73.129,
    71.797, 69.237, 68.674
]

anios = np.arange(1960, 2023)
decadas = ["Dcada 1960-1969", "Dcada 1970-1979", "Dcada 1980-1989", "Dcada 1990-1999", "Dcada 2000-2009", "Dcada 2010-2019", "Dcada 2020-2022"]
inicio_decadas = [1960, 1970, 1980, 1990, 2000, 2010, 2020]
fin_decadas = [1969, 1979, 1989, 1999, 2009, 2019, 2022]


medias_decada = []
for i in range(len(inicio_decadas)):
    anio_inicio = inicio_decadas[i]
    anio_fin = fin_decadas[i]
    indice_inicio = anio_inicio - 1960
    indice_fin = anio_fin - 1960 + 1  # +1 para incluir el ao final en el corte
    datos_decada = datos[indice_inicio:indice_fin]
    media = sum(datos_decada) / len(datos_decada)
    medias_decada.append(media)


# Define colores para cada dcada
colores = plt.cm.viridis(np.linspace(0.2, 0.8, len(decadas)))


# Crear el grfico de barras
plt.figure(figsize=(14, 7))
barras = plt.bar(decadas, medias_decada, color=colores)


# Aadir ttulos, etiquetas y grid
plt.title('Esperanza de vida promedio por dcada en Guatemala')
plt.xlabel('Dcada')
plt.ylabel('Esperanza de vida promedio')
plt.grid(axis='y', linestyle='-', alpha=0.7)
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()

print("Medias por dcada:")
for i, decada in enumerate(decadas):
    print(f"{decada}: {medias_decada[i]:.2f}")
```
